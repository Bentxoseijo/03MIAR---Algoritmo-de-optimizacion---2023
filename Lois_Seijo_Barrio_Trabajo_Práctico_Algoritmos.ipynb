### El espacio de soluciones
El espacio de soluciones vendrá dado por una lista de listas. Cada sublista estará formada por 3 elementos:

*   Posición 0: una tupla con los nombres de cada equipo (str).
*   Posición 1: la audiencia que atrae el partido (float).
*   Posición 2: el día y la hora que se le asigna (str).<br>

Ejemplo de como se representaría un partido: [('Celta', 'RealMadrid'), 1.04, 'S18']


### Función objetivo

La función objetivo en este caso será la audiencia total de todos los partidos, la cual queremos maximizar. Es decir se busca la suma máxima posible de los segundos elementos de todas las sublistas. En el código se respresentará como la variable "sum_audiencias" dentro de la función "horario_solucion".

### Restricciones
1. La asignación audiencias (basadas en un sábado a las 20:00) se hará con la función "Asignar_audiencia" que llama a "Audiencia_optima". Esta última se basa en una serie de condicionales que devolverán la audiencia de un encuentro según la categoría de los equipos que jueguen.






```
def Audiencia_optima(local, visitante, Eq_por_cat):
    if local in Eq_por_cat[0]:
        if visitante in Eq_por_cat[0]:
            aud = 2
        elif visitante in Eq_por_cat[1]:
            aud = 1.3
        else:
            aud = 1
    elif local in Eq_por_cat[1]:
        if visitante in Eq_por_cat[0]:
            aud = 1.3
        elif visitante in Eq_por_cat[1]:
            aud = 0.9
        else:
            aud = 0.75
    else:
        if visitante in Eq_por_cat[0]:
            aud = 1
        elif visitante in Eq_por_cat[1]:
            aud = 0.75
        else:
            aud = 0.47
    return aud


def Asignar_audiencia(lista_partidos, lista_eq_por_cat):
    jornada = []
    for encuentro in lista_partidos:
        jornada.append([(encuentro[0], encuentro[1]), Audiencia_optima(encuentro[0], encuentro[1], lista_eq_por_cat)])
    return jornada
```


2.  El resto de restricciones se harán dentro de la función "horario_solucion",
que se divide en 3 partes:

```
def horario_solucion(list_emparejamientos, coef_coind, coef_horarios):
    # ----- 1) Asignación de horas -----
    obligatorios = ['V20', 'L20']
    no_obligatorios = ['S12', 'S16', 'S18', 'S20', 'D12', 'D16', 'D18', 'D20']
    todas_horas = ['V20', 'S12', 'S16', 'S18', 'S20', 'D12', 'D16', 'D18', 'D20', 'L20']
    id_ob = 0
    while id_ob < 2:
        id = random.randint(0, len(list_emparejamientos) - 1)
        if len(list_emparejamientos[id]) < 3:
            list_emparejamientos[id].append(obligatorios[id_ob])
            id_ob += 1

    for i in range(len(list_emparejamientos)):
        if len(list_emparejamientos[i]) > 2:
            pass
        else:
            id = random.randint(0, len(no_obligatorios) - 1)
            list_emparejamientos[i].append(no_obligatorios[id])

    # ----- 2) Contaje de coincidencias y correción de audiencia correspondiente -----
    for fecha in todas_horas:
        coincidencias = -1
        coincidencias_id = []
        for i in range(len(list_emparejamientos)):
            if list_emparejamientos[i][2] == fecha:
                coincidencias += 1
                coincidencias_id.append(i)
        if coincidencias >= 1:
            for id_correccion in coincidencias_id:
                list_emparejamientos[id_correccion][1] *= 1 - coef_coind[coincidencias]

    # ----- 3) Correción de audiencia por fecha -----
    sum_audiencias = 0
    for i in range(len(list_emparejamientos)):
        coef_horario_id = todas_horas.index(list_emparejamientos[i][2])
        list_emparejamientos[i][1] *= coef_horarios[coef_horario_id]
        sum_audiencias += list_emparejamientos[i][1]

    return sum_audiencias, list_emparejamientos
```

1)  Asignación de horas. Como se puede ver, primero se asignan los horarios del viernes y del lunes a las 20h una única vez cada uno de manera aleatoria. Se usa el bucle While para asegurarnos de que no se asignan por error V20 y L20 a un mismo encuentro. Luego se asignarán horarios de manera aleatoria al resto de encuentros.

2) Contaje de coincidencias y correción de audiencia correspondiente. Se contarán las veces que se repitan las horas de los partidos ('V20', 'S12', etc.) en los terceros elementos de cada sublista para posteriormente aplicar un coeficiciente de corrección a la audiencia de cada partido: <br>

*   Coefs_Coincidencia = [0, 0.25, 0.45, 0.6, 0.7, 0.75, 0.78, 0.8]

Siendo el primer elemento lo que se le resta a la audiencia, en porcentaje, si hay 0 coincidencias y el último elemento lo que se le resta a cada partido coincidente (-80%) si hay 7 con el mismo horario.

3) Correción de audiencia por fecha. Se utilizan 2 listas cuyos índices relacionan cada horario a un coeficiente:


* todas_horas=['V20', 'S12', 'S16', 'S18', 'S20', 'D12', 'D16', 'D18', 'D20', 'L20']
*   Coefs_Horario=[0.4, 0.55, 0.7, 0.8, 1, 0.45, 0.75, 0.85, 1, 0.4]

Para obtener la cantidad de soluciones totales hay que tener en cuenta lo siguiente:

1.   Al principio, sin haber asignado ninguno de los 10 partidos, tedremos 10 posibles opciones para V20 ó L20.
2.   Luego, tedremos 9 posibles opciones para el otro de los 2 partidos mencionados.
3.   Después de esto, para el resto de partidos (quedan 8), habrá siempre 8 opciones para colocarlos: cualquier partido de sábado o domingo.

Dicho esto tendremos que el número total de soluciones es:

N_sol = 10 * 9 * 8^8 = 1509949440

```
#---DATOS INICIALES---

Coefs_Horario=[0.4,0.55,0.7,0.8,1,0.45,0.75,0.85,1,0.4]
Coefs_Coincidencia = [0, 0.25, 0.45, 0.6, 0.7, 0.75, 0.78, 0.8]

Lista_encuentros=[['Celta', 'RealMadrid'],
 ['Valencia', 'RealSociedad'],
 ['Mallorca', 'Eibar'],
 ['Athletic', 'Barcelona'],
 ['Leganés', 'Osasuna'],
 ['Villareal', 'Granada'],
 ['Alavés', 'Levante'],
 ['Espanyol', 'Sevilla'],
 ['Betis', 'Valladolid'],
 ['Atlético', 'Getafe']]

Equipos_por_categoria=[["RealMadrid", "RealSociedad", "Barcelona"], #A
                       ["Celta", "Valencia", "Athletic", "Villareal", "Alavés","Levante", "Espanyol", "Sevilla", "Betis", "Atlético", "Getafe"], #B
                       ["Mallorca", "Eibar", "Leganés", "Osasuna", "Granada", "Valladolid"]] #C
```

```
#---BÚSQUEDA ALEATORIA---

mejor_solucion = []
mejor_audiencia = 0

iters = 5000
for i in range(iters):
    partidos_sin_horario = Asignar_audiencia(Lista_encuentros, Equipos_por_categoria)
    audiencia_i, solucion_i = horario_solucion(partidos_sin_horario, Coefs_Coincidencia, Coefs_Horario)
    if audiencia_i > mejor_audiencia:
        mejor_solucion = solucion_i
        mejor_audiencia = audiencia_i

df = pd.DataFrame(mejor_solucion, columns=['Partido', 'Audiencia', 'Hora'])
display(df)
print('La audiencia de la mejor solución es: ' + str(round(mejor_audiencia, 4))+" Mill")
```
